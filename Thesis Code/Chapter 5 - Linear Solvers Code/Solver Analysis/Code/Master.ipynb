```python
import subprocess
#M max jobs 2979, working

def submit_jobs(start, end):
    """
    Submits job files from job{start}.pbs to job{end}.pbs using qsub command.
    
    :param start: The starting index of the job files.
    :param end: The ending index of the job files.
    """
    for i in range(start, end + 1):
        command = f"qsub Sjob{i}.pbs"
        subprocess.run(command, shell=True)

# Example usage: submit jobs 1 through 10
submit_jobs(251, 500)
```

```python
import os
import shutil

# Define the source and destination directories
source_directory = os.getcwd()
destination_directory = 'os.getcwd()' + '/batch1pbs'

# Ensure the destination directory exists
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Loop through the files in the source directory and move them if they contain 'job'
for filename in os.listdir(source_directory):
    if 'job' in filename:
        source_path = os.path.join(source_directory, filename)
        destination_path = os.path.join(destination_directory, filename)
        shutil.move(source_path, destination_path)
        print(f"Moved: {filename}")
```
